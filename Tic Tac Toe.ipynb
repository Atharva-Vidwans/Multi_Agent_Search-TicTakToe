```python
import sys, pygame
import keyboard
from pygame.locals import*
import time
pygame.init()

# Player X is the Human 
# Player O is the AI
# Human plays the first turn

class TicTakToe:
    
    def __init__(self):
        self.initialize_game()
    
    ####-----------------------------------------------------------------------------
    
    def initialize_game(self):
        self.current_state = [['.','.','.'],['.','.','.'],['.','.','.']]

        # Human starts with X
        self.player_turn = 'X'
        
    
    ####-----------------------------------------------------------------------------
    
    def check_if_valid(self, px, py):
        if px < 0 or px > 2 or py < 0 or py > 2:
            return False
        elif self.current_state[px][py] != '.':
            return False
        else:
            return True
    
    ####-----------------------------------------------------------------------------
    
    def check_if_end(self):
        # Vertical win
        for i in range(0, 3):
            if (self.current_state[0][i] != '.' and
                self.current_state[0][i] == self.current_state[1][i] and
                self.current_state[1][i] == self.current_state[2][i]):
                return self.current_state[0][i]

        # Horizontal win
        for i in range(0, 3):
            if (self.current_state[i] == ['X', 'X', 'X']):
                return 'X'
            elif (self.current_state[i] == ['O', 'O', 'O']):
                return 'O'

        # Main diagonal win
        if (self.current_state[0][0] != '.' and
            self.current_state[0][0] == self.current_state[1][1] and
            self.current_state[0][0] == self.current_state[2][2]):
            return self.current_state[0][0]

        # Second diagonal win
        if (self.current_state[0][2] != '.' and
            self.current_state[0][2] == self.current_state[1][1] and
            self.current_state[0][2] == self.current_state[2][0]):
            return self.current_state[0][2]

        # Is whole board full?
        for i in range(0, 3):
            for j in range(0, 3):
                # There's an empty field, we continue the game
                if (self.current_state[i][j] == '.'):
                    return None

        # It's a tie!
        return '.'
    
    ####-----------------------------------------------------------------------------
    

    def max(self):

        
        maxv = -20

        px = None
        py = None

        result = self.check_if_end()

 
        if result == 'X':
            return (-10, 0, 0)
        elif result == 'O':
            return (10, 0, 0)
        elif result == '.':
            return (0, 0, 0)

        for i in range(0, 3):
            for j in range(0, 3):
                if self.current_state[i][j] == '.':
                    # On the empty field player 'O' makes a move and calls Min
                    # That's one branch of the game tree.
                    self.current_state[i][j] = 'O'
                    (m, min_i, min_j) = self.min()
                    # Fixing the maxv value if needed
                    if m > maxv:
                        maxv = m
                        px = i
                        py = j
                    # Setting back the field to empty
                    self.current_state[i][j] = '.'
        return (maxv, px, py)
    
    ####-----------------------------------------------------------------------------
    
    # Player 'X' is min, in this case human
    def min(self):



        minv = 20

        qx = None
        qy = None

        result = self.check_if_end()

        if result == 'X':
            return (-10, 0, 0)
        elif result == 'O':
            return (10, 0, 0)
        elif result == '.':
            return (0, 0, 0)

        for i in range(0, 3):
            for j in range(0, 3):
                if self.current_state[i][j] == '.':
                    self.current_state[i][j] = 'X'
                    (m, max_i, max_j) = self.max()
                    if m < minv:
                        minv = m
                        qx = i
                        qy = j
                    self.current_state[i][j] = '.'

        return (minv, qx, qy)
    
    ####-----------------------------------------------------------------------------
    
    def play(self):
        
        width=320
        height=320
        white=(255,255,255)
        black=(0,0,0)
        blue = (0,0,255)
        LEFT = 1
        RIGHT = 3
        
        screen=pygame.display.set_mode((width,height))
        screen.fill(white)    

        # Vertical Lines 
        pygame.draw.line(screen,black,(100,0),(100,320),10)  
        pygame.draw.line(screen,black,(210,0),(210,320),10)

        # Horizantal Lines
        pygame.draw.line(screen,black,(0,100),(320,100),10)  
        pygame.draw.line(screen,black,(0,210),(320,210),10)

        cross = pygame.transform.scale(pygame.image.load(r"C:\Data\Working folder\Notebook\Tic tak toe\x.jpg"),(95,95)) 
        circle = pygame.transform.scale(pygame.image.load(r"C:\Data\Working folder\Notebook\Tic tak toe\o.png"),(95,95))
        
#         font = pygame.font.SysFont("comicsans",80)
#         YouWon = font.render('YOU LOSE!!', True, blue)
#         IWon = font.render('I WON!!', True, blue)
#         Tie = font.render("IT'S A TIE!!", True, blue)
#         textRectO = YouWon.get_rect()
#         textRectO.center = (width // 2, height // 2)
#         textRectX = IWon.get_rect()
#         textRectX.center = (width // 2, height // 2)
#         textRectT = Tie.get_rect()
#         textRectT.center = (width // 2, height // 2)

        
        pygame.display.update()
        done = True
        while done:
            
            #self.draw_board()
            self.result = self.check_if_end()

            # Printing the appropriate message if the game has ended
            if self.result != None:
                if self.result == 'X':
                    print('The winner is X!')
                    
                    pygame.display.update()
                elif self.result == 'O':
                    print('The winner is O!')
                    
                    pygame.display.update()
                elif self.result == '.':
                    print("It's a tie!")
                    
                    pygame.display.update()

                self.initialize_game()
                return

            # If it's player's turn
            if self.player_turn == 'X':

                while True:

                    #start = time.time()
                    #(m, qx, qy) = self.min()
                    #end = time.time()
                    #print('Evaluation time: {}s'.format(round(end - start, 7)))
                    #print('Recommended move: X = {}, Y = {}'.format(qx, qy))
                    for events in pygame.event.get():
                        #print(events)
                        if events.type == QUIT:
                            pygame.quit()
                            #sys.exit(0)
                        if events.type == pygame.MOUSEBUTTONDOWN and events.button == LEFT: 
                            x,y = events.pos
                            #print(x,y)
                            #### first row ####        
                            if x<100 and y <100 : 
                                px = 0
                                py = 0
                            if x>110 and x<210 and y<100 :
                                px = 1
                                py = 0
                            if x>220 and y<100 :
                                px = 2
                                py = 0                
                            #### second row ####
                            if x<100 and y>110 and y<210 :
                                px = 0
                                py = 1
                            if x>110 and x<210 and y>110 and y<210 :
                                px = 1
                                py = 1
                            if x>220 and y>110 and y<210 :
                                px = 2
                                py = 1                
                            #### third row ####
                            if x<100 and y>220 :
                                px = 0
                                py = 2
                            if x>110 and x<210 and y>220 :
                                px = 1
                                py = 2
                            if x>220 and y>220 :
                                px = 2
                                py = 2  

                            pygame.display.update()    
                            #px = int(input('Insert the X coordinate: '))
                            #py = int(input('Insert the Y coordinate: '))

                            (qx, qy) = (px, py)

                            if self.check_if_valid(px, py):
                                self.current_state[px][py] = 'X'
                                #### first row ####
                                if px==0 and py==0 : 
                                    screen.blit(cross, (0, 0))
                                if px==1 and py==0 :
                                    screen.blit(cross, (110, 0))
                                if px == 2 and py == 0 :
                                    screen.blit(cross, (220, 0))                
                                #### second row ####
                                if px == 0 and py == 1 :
                                    screen.blit(cross, (0, 110))
                                if px == 1 and py == 1 :
                                    screen.blit(cross, (110, 110))
                                if px == 2 and py == 1:
                                    screen.blit(cross, (220, 110))                
                                #### third row ####
                                if px == 0 and py == 2 :
                                    screen.blit(cross, (0, 220))
                                if px == 1 and py == 2 :
                                    screen.blit(cross, (110, 220))
                                if px == 2 and py == 2 :
                                    screen.blit(cross, (220, 220)) 
                                    
                                pygame.display.update()
                                self.player_turn = 'O'
                                #print("X over")
                                break
                            else:
                                
                                print('The move is not valid! Try again.')
                            break
                        break
                    break

            # If it's AI's turn
            else:
                (m, px, py) = self.max()
                self.current_state[px][py] = 'O'
                #### first row ####
                if px==0 and py==0 : 
                    screen.blit(circle, (0, 0))
                if px==1 and py==0 :
                    screen.blit(circle, (110, 0))
                if px == 2 and py == 0 :
                    screen.blit(circle, (220, 0))                
                #### second row ####
                if px == 0 and py == 1 :
                    screen.blit(circle, (0, 110))
                if px == 1 and py == 1 :
                    screen.blit(circle, (110, 110))
                if px == 2 and py == 1 :
                    screen.blit(circle, (220, 110))                
                #### third row ####
                if px == 0 and py == 2 :
                    screen.blit(circle, (0, 220))
                if px == 1 and py == 2 :
                    screen.blit(circle, (110, 220))
                if px == 2 and py == 2 :
                    screen.blit(circle, (220, 220)) 
                self.player_turn = 'X'
                
                pygame.display.update()

        for events in pygame.event.get():
            #print(events)
            if events.type == QUIT:
                done = False
                pygame.quit()
                
    

game = TicTakToe()
game.play()


   
```
